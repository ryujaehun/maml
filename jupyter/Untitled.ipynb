```python
from torchmeta.datasets.helpers import omniglot
from torchmeta.utils.data import BatchMetaDataLoader
import matplotlib.pyplot as plt
import torch.nn.functional as F
from collections import namedtuple
from torchmeta.datasets import Omniglot, MiniImagenet
from torchmeta.toy import Sinusoid
from torchmeta.transforms import ClassSplitter, Categorical, Rotation
from torchvision.transforms import ToTensor, Resize, Compose
from maml.model import ModelConvOmniglot, ModelConvMiniImagenet, ModelMLPSinusoid
from maml.utils import ToTensor1D
import torchvision.transforms as transforms


transform = Compose([Resize(28), ToTensor()])
dataset_transform = ClassSplitter(shuffle=True,
                                      num_train_per_class=7,
                                      num_test_per_class=3)
dataset = Omniglot("data", transform=transform,
                                      target_transform=Categorical(5),
                                      num_classes_per_task=5,
                                      meta_train=True,
                                      dataset_transform=dataset_transform,
                                      download=True)
dataloader = BatchMetaDataLoader(dataset, batch_size=2, num_workers=4)

for batch in dataloader:
    train_inputs, train_targets = batch["train"]
    print('Train inputs shape: {0}'.format(train_inputs.shape))    # (16, 25, 1, 28, 28)
    print('Train targets shape: {0}'.format(train_targets.shape))  # (16, 25)

    test_inputs, test_targets = batch["test"]
    print('Test inputs shape: {0}'.format(test_inputs.shape))      # (16, 75, 1, 28, 28)
    print('Test targets shape: {0}'.format(test_targets.shape))    # (16, 75)
    for outer in train_targets:
        print(outer)
        #for inner in outer:
            #print(inner)
#             plt.imshow(transforms.ToPILImage()(outer[10]))
#             plt.show()
        break
        print('exit')
    for outer in test_targets:
        print(outer)
#         for inner in outer:
#             plt.imshow(transforms.ToPILImage()(outer[10]))
#             plt.show()
        break
        print('exit')
    break
```
